1.Write a Python Program to Check if a Number is Positive, Negative or Zero?
---------------------------

```python
user_number=int(input("Enter Any Number"))
if user_number==0:
    print("{}  is Zero Number".format(user_number))
elif user_number>0:
    print("{}  is Positive Number".format(user_number))
else :
    print("{}  is Negative Number".format(user_number))
```

2.Write a Python Program to Check if a Number is Odd or Even?
----------------------------

```python
user_number=int(input("Enter Any Number"))
if user_number%2==0:
    print("{}  is Even Number".format(user_number))
else :
    print("{}  is Odd Number".format(user_number))
```

3.Write a Python Program to Check Leap Year?
-------------------------

```python
import calendar as cd
user_input=int(input("Enter Year"))
def LeapYear(user_input):
    return cd.isleap(user_input)
if LeapYear(user_input):
    print("{} is Leap Year".format(user_input))
else:
    print("{} is not Leap Year".format(user_input))
```

-----------------
